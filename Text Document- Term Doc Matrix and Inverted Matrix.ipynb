```python
import numpy as np
```

#### Here, we are using three text files containing news-pieces about the Euro 2020 football tournament.

```python
# Reading the text files
file1 = open(r"C:\Users\user\Documents\Data Science\IRT\Euro1.txt")
file2 = open(r"C:\Users\user\Documents\Data Science\IRT\Euro2.txt")
file3 = open(r"C:\Users\user\Documents\Data Science\IRT\Euro3.txt")
```

```python
# Python file method readlines() reads until EOF using readline() and returns a list containing the lines.
x=file1.readlines()
y=file2.readlines()
z=file3.readlines()
```

```python
docs=x+y+z
docs
```

```python
# Gather the set of all unique terms
unique_terms = {term for doc in docs for term in doc.split()}
unique_terms
```

## Term Document Matrix

```python
# Construct a term-document matrix

doc_term_matrix = {}

for term in unique_terms:
    doc_term_matrix[term] = []
    
    for doc in docs:
        if term in doc:
            doc_term_matrix[term].append(1)
        else: doc_term_matrix[term].append(0)

doc_term_matrix
```

```python
# The query to find all documents containing "home" AND "Kane"AND "Italy" AND "penalty"
# Is just a bitwise AND:



docs_array = np.array(docs, dtype='object')

v1 = np.array(doc_term_matrix['Kane'])    
v2 = np.array(doc_term_matrix['penalty'])
v3 = np.array(doc_term_matrix['Italy'])
print(v1)
print(v2)
print(v3)
print('-------')
v = v1 & v2 & v3
print(v)
```

```python
# We can now get the matching documents from our corpus with the result. Here, the result is the third file.
[doc for doc in v * docs_array if doc]
```

```python
# Bitwise OR to construct 'football' or 'goal' queries.

v1 = np.array(doc_term_matrix['football'])    
v2 = np.array(doc_term_matrix['goal'])

print(v1)
print(v2)
print('-------')
v = v1 | v2
print(v)
```

```python
# We can now get the matching documents from our corpus with the result. Here, terms like football or goal can easily be found in any football article. Hence, as we can see, all three files contain the result.
[doc for doc in v * docs_array if doc]
```

## Inverted Matrix

```python
# Construct an inverted index

inverted_index = {}

for i, doc in enumerate(docs):
    for term in doc.split():
        if term in inverted_index:
            inverted_index[term].add(i)
        else: inverted_index[term] = {i}

inverted_index
```

```python
# Now we can search for any term
posting_list = inverted_index['Euro']
posting_list
```

We can define functions to find our required terms across the documents for OR condition and AND condition

```python
def or_postings(posting1, posting2):
    p1 = 0
    p2 = 0
    result = list()
    while p1 < len(posting1) and p2 < len(posting2):
        if posting1[p1] == posting2[p2]:
            result.append(posting1[p1])
            p1 += 1
            p2 += 1
        elif posting1[p1] > posting2[p2]:
            result.append(posting2[p2])
            p2 += 1
        else:
            result.append(posting1[p1])
            p1 += 1
    while p1 < len(posting1):
        result.append(posting1[p1])
        p1 += 1
    while p2 < len(posting2):
        result.append(posting2[p2])
        p2 += 1
    return result
```

```python
pl_1 = list(inverted_index['football'])
pl_2 = list(inverted_index['final'])
or_postings(pl_1, pl_2)
```

The terms 'football' OR 'final' exists in all three files.

```python
def and_postings(posting1, posting2):
    p1 = 0
    p2 = 0
    result = list()
    while p1 < len(posting1) and p2 < len(posting2):
        if posting1[p1] == posting2[p2]:
            result.append(posting1[p1])
            p1 += 1
            p2 += 1
        elif posting1[p1] > posting2[p2]:
            p2 += 1
        else:
            p1 += 1
    return result
```

```python
pl_1 = list(inverted_index['Kane'])
pl_2 = list(inverted_index['scored'])
and_postings(pl_1, pl_2)
```

The terms 'Kane' AND 'scored' exist only in the first file.
